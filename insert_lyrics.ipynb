```python
import mysql.connector
import pandas as pd

# Cấu hình kết nối database
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': '',
    'database': 'momudb'
}

# Kết nối MySQL và lấy dữ liệu từ bảng songs
def export_songs_to_excel():
    try:
        connection = mysql.connector.connect(**db_config)
        query = """
            SELECT
                id,
                title,
                explicit,
                album_cover,
                popularity,
                preview_url,
                release_date,
                is_playable,
                duration_ms,
                album_id,
                url,
                lyrics
            FROM songs
        """
        df = pd.read_sql(query, con=connection)

        # Ghi ra file Excel
        df.to_excel("songs_export.xlsx", index=False)
        print("✅ Xuất thành công: songs_export.xlsx")

    except mysql.connector.Error as e:
        print("❌ Lỗi khi kết nối MySQL:", e)
    finally:
        if connection.is_connected():
            connection.close()

# Gọi hàm
export_songs_to_excel()
```
